```python
#小明想要存錢買一輛機車,機車每輛30000元，他將每月存的錢輸入，當存款足夠買機車時，就顯示提示訊息告知。
deposite =0
num =0
while(deposite<30000):
    num +=1
    input_value =int(input("請輸入第{0}月份存款:".format(num)))
    deposite += input_value
print("恭喜!達成存款目標存了%d個月,總存款為%d元"%(num,deposite))
```

```python
#小美是一位教師，請你以while迴圈方式為小美設計一個輸入成績的程式，如果輸入負數表示成績輸入結束，在輸入成績結束後顯示班上總成績及平均成績。

sum=0
num=0
input_score=int(input(f"請輸入第{num+1}位學生成績:"))
while(input_score>=0):
  sum+=input_score
  num+=1
  input_score=int(input(f"請輸入第{num+1}位學生成績:")) 
print(f"學生總分為{sum}")
print(f"學生平均為{sum/num}")
```

```python
import random
min=1
max=100
random_num=random.randint(min,max)
used_num=0

while(True):
    keyin_value=int(input("猜數字範圍{0}~{1}:".format(min,max)))
    used_num+=1
    if(keyin_value>=min and keyin_value<=max):
        if keyin_value==random_num:
            print("賓果!您猜對了,答案是:%d" % random_num)
            print(f"您猜了{used_num}次")
            break
        elif(keyin_value>random_num):
            print("再小一點")
            max=keyin_value-1
        elif(keyin_value<random_num):
            print("再大一點")
            min=keyin_value+1

        print(f"您已猜了{used_num}次")
    else:
      print("超出範圍了!")
      continue
print("Game Over")
```
